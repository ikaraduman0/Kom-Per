###  Bir pizza restoran yeni eit pizzalar hazrlamak istemektedir. Hazrlanacak pizzalar ile ilgili unlar bilinmektedir.

    > Yeni pizzalarda 8 eit temel malzeme kullanlacaktr. (Peynir, Zeytin, Msr, Domates, Biber, Mantar, Salam, Tavuk)
    > Her pizza bu temel malzemelerden 6'snn bir araya gelmesinden oluacaktr.
    > Salam ve Tavuk ayn pizza ierisinde kullanlmayacaktr.

### Bu durumda restorant mterisine ka farkl pizza sunabilir? 


# zm

 Kombinasyon formlmz (n)! / ((n-r)!.r!) dir. 
 
 Bu durumda 8'in 6'l kombinasyonundan, 6 malzemeli pizza says 8! / ((8-6)!.6! = 8.7 / 2.1 = 28 dir.
 
 Salam ve tavuun bir arada bulunduu pizza saysn bulmak iin 
 Salam ve tavuu ayr bir grup, dier malzemeleride ayr bir grup olarak deerlendirdiimizde,
 Salam ve tavuk iin 2'nin 2'li kombinasyonu kullanlr, salam ve tavuun olmad malzemelerden oluan 2. grupta geriye  
 kalan malzeme says 8-2=6 ve pizzada kullanlabilecek malzeme says 6-2= 4 tr. Bu durumda 2. grup iin 6'nn 4'l 
 kombinasyonu kullanlr. 
  
 (2! / (2-2)!.2!)=1
 (6! / ((6-4)!.4!))=15
 Salam ve tavuun bir arada bulunduu pizza says 1 x 15 =15 olur.
  
 Salam ve tavuun bir arada bulunmad toplam 28-15 = 13 pizza mteriye sunulabilir. 

```python
import itertools  

# nce malzemenin tanm
malzemeler=["Peynir", "Zeytin", "Msr", "Domates", "Biber", "Mantar", "Salam", "Tavuk"]

pizzalar = list()
for p in list(itertools.combinations(malzemeler, 6)):     
    if not(("Tavuk" in p) and ("Salam" in p)):        
        pizzalar.append(p)

# Pizza eitlerinin ekrana yazdrlmas
for m in pizzalar: print(m)

print("\n Hesaplanan pizza says",len(pizzalar),"\n")
```

### 5 Futbol takmnn yer ald deplasmanl bir futbol turnuvasnda gerekleecek takm elemeleri nelerdir?

# zm 

Turnuvada toplam 5 (n) takm var. Takmlar kendileri ile oynamayamazlar ve her ma 2 (r) takmn arasnda gerekleir. Bu durumda Permtasyon n! / (n-r)! formlnden, Toplam = 5! / (5-2)! = 5.4 = 20 karlama sz konusu olacaktr.

```python
import itertools  

# nce takmlarn tanm
takimlar=["Galatasaray","Fenerbahe","Beikta","Trabzonspor","Kocaelispor"]

# Hazr fonksiyon ile hesaplama
maclar=list(itertools.permutations(takimlar,2))

# Mac permtasyonlarnn ekrana yazdrlmas
for m in maclar: print(m)

print("\n Permutasyon fonksiyonu ile hesaplanan ma says",len(maclar),"\n")
```
