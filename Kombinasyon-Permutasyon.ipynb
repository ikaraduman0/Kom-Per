# Soru - 1

###  Bir pizza restoranı yeni çeşit pizzalar hazırlamak istemektedir. Hazırlanacak pizzalar ile ilgili şunlar bilinmektedir.

    > Yeni pizzalarda 8 çeşit temel malzeme kullanılacaktır. (Peynir, Zeytin, Mısır, Domates, Biber, Mantar, Salam, Tavuk)
    > Her pizza bu temel malzemelerden 6'sının bir araya gelmesinden oluşacaktır.
    > Salam ve Tavuk aynı pizza içerisinde kullanılmayacaktır.

### Bu durumda restorant müşterisine kaç farklı pizza sunabilir? 


# Çözüm

 Kombinasyon formülümüz (n)! / ((n-r)!.r!) dir. 
 
 Bu durumda 8'in 6'lı kombinasyonundan, 6 malzemeli pizza sayısı 8! / ((8-6)!.6! = 8.7 / 2.1 = 28 dir.
 
 Salam ve tavuğun bir arada bulunduğu pizza sayısını bulmak için 
 Salam ve tavuğu ayrı bir grup, diğer malzemeleride ayrı bir grup olarak değerlendirdiğimizde,
 Salam ve tavuk için 2'nin 2'li kombinasyonu kullanılır, salam ve tavuğun olmadığı malzemelerden oluşan 2. grupta geriye  
 kalan malzeme sayısı 8-2=6 ve pizzada kullanılabilecek malzeme sayısı 6-2= 4 tür. Bu durumda 2. grup için 6'nın 4'lü 
 kombinasyonu kullanılır. 
  
 (2! / (2-2)!.2!)=1
 (6! / ((6-4)!.4!))=15
 Salam ve tavuğun bir arada bulunduğu pizza sayısı 1 x 15 =15 olur.
  
 Salam ve tavuğun bir arada bulunmadığı toplam 28-15 = 13 pizza müşteriye sunulabilir. 

```python
import itertools  

# Önce malzemenin tanımı
malzemeler=["Peynir", "Zeytin", "Mısır", "Domates", "Biber", "Mantar", "Salam", "Tavuk"]

pizzalar = list()
for p in list(itertools.combinations(malzemeler, 6)):     
    if not(("Tavuk" in p) and ("Salam" in p)):        
        pizzalar.append(p)

# Pizza çeşitlerinin ekrana yazdırılması
for m in pizzalar: print(m)

print("\n Hesaplanan pizza sayısı",len(pizzalar),"\n")
```

# Soru - 2

### 5 Futbol takımının yer aldığı deplasmanlı bir futbol turnuvasında gerçekleşecek takım eşleşmeleri nelerdir?

# Çözüm 

Turnuvada toplam 5 (n) takım var. Takımlar kendileri ile oynamayamazlar ve her maç 2 (r) takımın arasında gerçekleşir. Bu durumda Permütasyon n! / (n-r)! formülünden, Toplam = 5! / (5-2)! = 5.4 = 20 karşılaşma söz konusu olacaktır.

```python
import itertools  

# Önce takımların tanımı
takimlar=["Galatasaray","Fenerbahçe","Beşiktaş","Trabzonspor","Kocaelispor"]

# Hazır fonksiyon ile hesaplama
maclar=list(itertools.permutations(takimlar,2))

# Mac permütasyonlarının ekrana yazdırılması
for m in maclar: print(m)

print("\n Permutasyon fonksiyonu ile hesaplanan maç sayısı",len(maclar),"\n")
```

```python
# Önce takımların tanımı
takimlar=["Galatasaray","Fenerbahçe","Beşiktaş","Trabzonspor","Kocaelispor"]

# Hazır fonksiyon olmadan hesaplama
yenimaclar=[(t,r) for t in takimlar for r in takimlar if (r!=t)]

# Mac permütasyonlarının ekrana yazdırılması
for y in yenimaclar: print(y)

print("\nManuel hesaplanan maç sayısı",len(yenimaclar))
```
